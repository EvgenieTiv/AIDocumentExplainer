```python
from pathlib import Path

from load_data import (
    extract_text_from_pdf,
    preview_pdf_pages,
    clean_text_for_llm,
    truncate_text,
    remove_image_captions
)
from process_model import summarize_document_to_json, print_summary

# =====================
# Configuration
# =====================
pdf_path = Path("Data/haiti_gangs.pdf")
# document_mode = "research"   # later: news_article, policy, general, etc.
document_mode = "news_article"
preview_pages = True
preview_chars_per_page = 1200
max_chars_for_llm = 12000


# =====================
# Step 1. Load PDF
# =====================
full_text, page_count = extract_text_from_pdf(pdf_path)

# print(f"PDF loaded: {pdf_path.name}")
# print(f"Page count: {page_count}")
# print(f"Raw text length: {len(full_text)}")


# =====================
# Step 2. Optional page preview
# =====================
if preview_pages:
    preview_pdf_pages(pdf_path, chars_per_page=preview_chars_per_page)


# =====================
# Step 3. Clean text
# =====================
cleaned_text = clean_text_for_llm(full_text)
cleaned_text = remove_image_captions(cleaned_text)

# print(f"Cleaned text length: {len(cleaned_text)}")
# print("\nCleaned text preview:\n")
# print(cleaned_text[:1500])


# =====================
# Step 4. Truncate for LLM
# =====================
truncated_text = truncate_text(cleaned_text, max_chars=max_chars_for_llm)

# print(f"\nFinal text length for LLM: {len(truncated_text)}")
# print("\nFinal text preview:\n")
# print(truncated_text[:1500])


# =====================
# Step 5. Run model
# =====================
result = summarize_document_to_json(document_mode, truncated_text)


# =====================
# Step 6. Print result
# =====================
# print("\n" + "=" * 80)
# print("DOCUMENT EXPLANATION RESULT")
# print("=" * 80 + "\n")

print_summary(result)
```
